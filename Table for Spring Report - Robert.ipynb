```python
# get the data into a more presentable table.



def render_mpl_table(data, col_width=3.0, row_height=0.700, font_size=14,
                     header_color='#ffb07c', row_colors=['#f1f1f2', 'w'], edge_color='w',
                     bbox=[0, 0, 1, 1], header_columns=0,
                     ax=None, **kwargs):
    if ax is None:
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height])
        fig, ax = plt.subplots(figsize=size)
        ax.axis('off')
    mpl_table = ax.table(cellText=data.values, bbox=bbox, colLabels=data.columns, **kwargs)
    mpl_table.auto_set_font_size(True)
    mpl_table.set_fontsize(font_size)
    

    for k, cell in mpl_table._cells.items():
        cell.set_edgecolor(edge_color)
        if k[0] == 0 or k[1] < header_columns:
            cell.set_text_props(weight='bold', color='k')
            cell.set_facecolor(header_color)
        else:
            cell.set_facecolor(row_colors[k[0]%len(row_colors) ])
    return ax.get_figure(), ax

fig,ax = render_mpl_table(df, header_columns=0, col_width=6.5)
fig.savefig("TableSpring21.png")

fig.show()
```

```python
# seasonal vs annual average discharge difference plot.
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = [1952,1955,1956,1959,1969,1979,1981,1985,1986,1989,1994,1996,1997,1998,1999,2003,
          2004,2006,2008,2018]
annual_mean = [224.9,168.2,95.0,148.8,159.9,354.7,81.0,217.9,123.9,260.5,232.6,320.7,
               181.1,306.5,124.2,413.3,269.2,191.7,116.2,350.8]
seas_mean = [219.4,212.8,86.0,135.7,197.4,318.0,88.0,127.2,106.4,299.4,246.2,
              260.5,175.0,282.4,110.5,438.5,235.3,122.4,133.1,301.7]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars
#width = np.diff(labels).min()

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, annual_mean, width, label='Annual Average Discharge')
rects2 = ax.bar(x + width/2, seas_mean, width, label='Seasonal Average Discharge')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Discharge (cms)')
ax.set_title('Year')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.autofmt_xdate()
fig.set_figheight(20)
fig.set_figwidth(32.36)
fig.savefig("DischargeCompare21.png")
plt.show()
```
