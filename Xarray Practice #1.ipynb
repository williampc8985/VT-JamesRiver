```python
#xarray practice
#customary imports
import numpy as np
import pandas as pd
import xarray as xr
```

```python
np.random.seed(123)
size = 4
temp = 15 + 10* np.random.randn(size)
lat = np.random.uniform(low = -90, high = 90,size = size)
lon = np.random.uniform(low = -180, high = 180, size = size)
#round to 2 digits after decimal
temp, lat, lon, = np.around([temp, lat, lon], decimals = 2)
```

```python
# how to represent this data in Pandas
# 1- dataframe
df = pd.DataFrame({'temp':temp, 'lat':lat, 'lon': lon})
df
```

```python
# now we can create a data array from this. There are four ways to do this.
#1. Create a data array from a series, with a two-level index = idx
idx = pd.MultiIndex.from_arrays(arrays = [lat,lon],names = ['lat','lon'])
s = pd.Series(data = temp, index = idx)
s
# use from_series method
da = xr.DataArray.from_series(s)
da
```

```python
df_pv = df.pivot(index = 'lat', columns = 'lon')
#drop the first of columns as its not necessary
df_pv = df_pv.droplevel(0, axis = 1)
df_pv
```

```python
# now the data has been pivoted, and a data array can be produced
da = xr.DataArray(data = df_pv)
da
```

```python
#create a data array usign a constructor
#This is the way to construct a data array manually

# get pivoted data as 2-d array (4,4)

temp_2d = df_pv.values
da = xr.DataArray(data = temp_2d, dims = ['lat','lon'], coords = [lat,lon])
da
```

```python
np.random.seed(123)
shape = (1,4) #needs to be 2d could be (2,2) or (4,1)
# all three arrays have the same shape
temp = 15+10*np.random.randn(*shape)
lat = np.random.uniform(low = -90, high = 90, size = shape)
lon = np.random.uniform(low = -180, high = 180, size = shape)
#round to two digits after decimal
temp, lat, lon = np.around([temp, lat,lon],decimals = 2)
```

```python
da = xr.DataArray(data = temp,coords={'lat':(['x','y'],lat),'lon': (['x','y'],lon)},dims = ['x','y'])
```
